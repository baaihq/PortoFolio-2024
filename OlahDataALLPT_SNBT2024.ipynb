```python
from google.colab import files
import pandas as pd
```

```python
# Memanggil file CSV dari github.
url = 'https://raw.githubusercontent.com/baaihq/PortoFolio-2024/main/AllUnivPoli-2024.csv'
df = pd.read_csv(url)
```

```python
# Memanggil DataFrame df
df
```

```python
# Mengecek tipe data dari DataFrame
df.dtypes
```

```python
# Merubah tipe data kolom ke-2 menjadi string.
df.iloc[:, 1] = df.iloc[:, 1].astype(str)

# Merubah tipe data kolom ke-4 dan ke-5 menjadi integer.
df.iloc[:, 4] = df.iloc[:, 4].astype(int)
df.iloc[:, 5] = df.iloc[:, 5].astype(int)
```

```python
# Mengecek kembali tipe data dari DataFrame
df.dtypes
```

```python
# Hitung Rasio Ketetatan (Daya Tampung / Peminat).
df['Rasio Keketetatan'] = df.apply(lambda row: row.iloc[4] / row.iloc[5] if row.iloc[5] != 0 else 0, axis=1) #Jika ada pembagi NOL, maka nilai 'Rasio Ketetatan' akan dituliskan menjadi NOL.
```

```python
# Hitung Presentase (%) dari Rasio Ketetatan.
df['Presentase (%)'] = (df['Rasio Keketetatan'] * 100)
```

```python
# Melihat tampilan baru dari DataFrame.
df
```

```python
# Simpan DataFrame ke file CSV
df.to_csv(r'PTN-PTKIN-PTVOKASI.csv', index = False)
```
